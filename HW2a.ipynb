```python
import datetime
import unittest

# Brand function
def my_brand(hw_assignment):
    name = "Karthik Darapaneni"
    course = "Course 2024F-SSW567-A"
    pattern = "====%s===="
    now = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")

    print(pattern % name)
    print(pattern % course)
    print(pattern % hw_assignment)
    print(pattern % now)

# Test the brand function
my_brand('Hw 01')
print('')

# Classify triangle function
def classify_triangle(a, b, c):
    # Ensure inputs are valid lengths for triangle sides
    if a <= 0 or b <= 0 or c <= 0 or (a + b <= c) or (a + c <= b) or (b + c <= a):
        return "Invalid"
    
    # Check for right triangle using Pythagorean theorem
    right_triangle = (
        round(a**2 + b**2, 5) == round(c**2, 5) or
        round(a**2 + c**2, 5) == round(b**2, 5) or
        round(b**2 + c**2, 5) == round(a**2, 5)
    )
    
    # Check for triangle types
    if a == b == c:
        return "Equilateral"
    elif a == b or b == c or a == c:
        return "Isosceles and Right" if right_triangle else "Isosceles"
    else:
        return "Scalene and Right" if right_triangle else "Scalene"

# Test cases for the classify_triangle function
class TestTriangleClassification(unittest.TestCase):
    def test_triangle(self):
        test_cases = [
            (5, 5, 5),
            (3, 3, 4),  
            (3, 3, 4.24),  
            (6, 8, 10),  
            (25, 25, 25),
            (1, 2, 3),
            (-1, 2, 3),
            (7, 9, 12)  
        ]
        
        for a, b, c in test_cases:
            result = classify_triangle(a, b, c)
            print(f"classify_triangle({a}, {b}, {c}) => Result: {result}")

# Run the unit tests in Jupyter Notebook
def run_tests():
    suite = unittest.TestLoader().loadTestsFromTestCase(TestTriangleClassification)
    unittest.TextTestRunner(verbosity=2).run(suite)

# Run the tests
run_tests()
```

```python
import unittest

# Define the classifyTriangle function
def classifyTriangle(a, b, c):
    # Sort the sides to easily apply the triangle conditions
    sides = sorted([a, b, c])
    a, b, c = sides  # a is the smallest, c is the largest

    # Check if it's not a triangle
    if a + b <= c:
        return 'NotATriangle'
    
    # Check for equilateral
    if a == b == c:
        return 'Equilateral'
    
    # Check for right triangle using Pythagorean theorem
    if a**2 + b**2 == c**2:
        return 'Right'
    
    # Check for isosceles
    if a == b or b == c:
        return 'Isosceles'
    
    # If none of the above, it must be scalene
    return 'Scalene'

# Test class for classifying triangles
class TestTriangles(unittest.TestCase):
    def testRightTriangle1(self): 
        self.assertEqual(classifyTriangle(3, 4, 5), 'Right', '3,4,5 is a Right triangle')

    def testRightTriangle2(self): 
        self.assertEqual(classifyTriangle(5, 3, 4), 'Right', '5,3,4 is a Right triangle')
        
    def testRightTriangle3(self): 
        self.assertEqual(classifyTriangle(8, 6, 10), 'Right', '8,6,10 is a Right triangle')
    
    def testRightTriangle4(self): 
        self.assertNotEqual(classifyTriangle(21, 6, 10), 'Right', '21,6,10 is a Not Right triangle')

    # Testing Equilateral Triangles
    def testEquilateralTriangle1(self): 
        self.assertEqual(classifyTriangle(1, 1, 1), 'Equilateral', '1,1,1 should be equilateral')

    def testEquilateralTriangle2(self): 
        self.assertEqual(classifyTriangle(30, 30, 30), 'Equilateral', '30,30,30 should be equilateral')

    # Testing Scalene Triangles
    def testScaleneTriangle1(self):
        self.assertEqual(classifyTriangle(9, 10, 11), 'Scalene', '9,10,11 should be scalene')

    # Testing Not a Triangle
    def testNotATriangle1(self):
        self.assertEqual(classifyTriangle(2, 2, 8), 'NotATriangle', '2,2,8 should not be a triangle')

# Run the tests in Jupyter Notebook
def run_tests():
    suite = unittest.TestLoader().loadTestsFromTestCase(TestTriangles)
    unittest.TextTestRunner(verbosity=2).run(suite)

# Call the function to run the tests
run_tests()
```
